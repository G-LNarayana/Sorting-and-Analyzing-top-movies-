Dropping unwanted columns

```python
df.drop(columns=['id','imdb_id', 'homepage', 'cast','tagline','overview','budget_adj','revenue_adj'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(how='any', subset=['director','genres'], inplace=True)
```

Fill null cell with 0

```python
df['production_companies']=df['production_companies'].fillna(0)
df['keywords']=df['keywords'].fillna(0)
```

rounding decimal 2 values

```python
df ['popularity']=df['popularity'].round(2)


```

Insert profit roi columns

```python
df.insert(3,'profit',df.revenue-df.budget)
```

```python
df.insert(4,'roi',df.profit/df.budget)
```

```python
df['roi']=df['roi'].round(2)
```

```python
df
```

check roi null

```python
df.roi.value_counts()
```

test infinite value

```python
non_finite_values= ~ np.isfinite(df['roi'])
```

```python
non_finite_values.sum()
```

```python
df['roi']=df['roi'].replace([np.inf,-np.inf],np.nan)
```

Top 5 Highest Grossing Movies

```python
df10= df.groupby('original_title')['profit'].sum().sort_values(ascending=False).head(5)
df10
```

```python
explode_list=[0.05,0,0,0,0]
df10.plot(kind='pie',autopct='%1.1f%%', startangle = 90,explode= explode_list, colors= plt.cm.Paired.colors)
plt.title('Top 5 Highest Grossers')
plt.legend(labels = df10.index, loc = 'upper right')
plt.axis('equal')
plt.show()
```
